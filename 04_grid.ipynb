# grid()
 is used to show grid lines on the graph.

These lines help read values more easily.

```python
import matplotlib.pyplot as plt

x = [1, 2 ,3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.plot(x,y)
plt.grid(axis='x',
        color = "lightgray",
        linewidth = 2,
        linestyle = "dashed")
plt.show()
```
