# title() 
title() is used to add a heading/title to the graph.

```python
import matplotlib.pyplot as plt

x = [2021,2022,2023,2024,2025,2026]
y = [10,19,27,39,60,100]

plt.plot(x,y, marker = "o")
plt.title("No of Students per Year")
plt.show()
```

# xlabe() and ylabel()
| Function   | Work            |
| ---------- | --------------- |
| `xlabel()` | name for X-axis |
| `ylabel()` | name for Y-axis |


```python
import matplotlib.pyplot as plt

x = [2021,2022,2023,2024,2025,2026]
y = [10,19,27,39,60,100]

plt.plot(x,y, marker = "o")
plt.title("No of Students per Year")
plt.xlabel("Year")
plt.ylabel("Students")
plt.show()
```

# fontsize

```python
import matplotlib.pyplot as plt

x = [2021,2022,2023,2024,2025,2026]
y = [10,19,27,39,60,100]

plt.plot(x,y, marker = "o")
plt.title("No of Students per Year", fontsize = 20)
plt.xlabel("Year", fontsize = 20)
plt.ylabel("Students", fontsize = 20)
plt.show()
```

# family
family is used to set the font style/type of text.

```python
import matplotlib.pyplot as plt

x = [2021,2022,2023,2024,2025,2026]
y = [10,19,27,39,60,100]

plt.plot(x,y, marker = "o")
plt.title("No of Students per Year", 
          fontsize = 20,
          family = "monospace")
plt.xlabel("Year", fontsize = 20)
plt.ylabel("Students", fontsize = 20)
plt.show()
```

# fontweight

```python
import matplotlib.pyplot as plt

x = [2021,2022,2023,2024,2025,2026]
y = [10,19,27,39,60,100]

plt.plot(x,y, marker = "o")
plt.title("No of Students per Year", 
          fontsize = 20,
          family = "monospace",
          fontweight = "bold")
plt.xlabel("Year", fontsize = 20)
plt.ylabel("Students", fontsize = 20)
plt.show()
```

# color

```python
import matplotlib.pyplot as plt

x = [2021,2022,2023,2024,2025,2026]
y = [10,19,27,39,60,100]

plt.plot(x,y, marker = "o")
plt.title("No of Students per Year", 
          fontsize = 20,
          family = "monospace",
          color = "r")
plt.xlabel("Year", fontsize = 20)
plt.ylabel("Students", fontsize = 20)
plt.show()
```

# ticks

Ticks are the small marks and numbers on axes.

# tick_params()

It helps you change:

tick color

tick size

tick direction

label size

```python
import matplotlib.pyplot as plt

x = [2021,2022,2023,2024,2025,2026]
y = [10,19,27,39,60,100]

plt.plot(x,y, marker = "o")
plt.title("No of Students per Year", 
          fontsize = 20,
          fontweight = "bold")
plt.tick_params(axis="both",
                color = "#22a51e",
                width = 3,
                length = 5,
                labelcolor = "r")
plt.xlabel("Year", fontsize = 20)
plt.ylabel("Students", fontsize = 20)
plt.show()
```

# loc means location.

It is used to control where something appears.

```python
import matplotlib.pyplot as plt

x = [2021,2022,2023,2024,2025,2026]
y = [10,19,27,39,60,100]

plt.plot(x,y, marker = "o")
plt.title("No of Students per Year", 
          fontsize = 20,
          family = "monospace",
          color = "r",
          loc= "left")
plt.xlabel("Year", fontsize = 20,loc="left")
plt.ylabel("Students", fontsize = 20, loc="bottom")
plt.show()
```
