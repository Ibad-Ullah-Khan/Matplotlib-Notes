# bar()
bar() is used to create a bar chart.

A bar chart shows data using rectangular bars.

```python
import matplotlib.pyplot as plt

food = ["Rice", "Fruits", "Vegetables", "Meat", "Milk"]
amount = [3, 2, 4, 1, 2]

plt.bar(food, amount, color = "lightgreen", width = 0.8)
plt.title("Category Conusme daily", fontweight = "bold")
plt.xlabel("Food Category", fontweight = "bold")
plt.ylabel("Amount Conusment", fontweight = "bold")
plt.show()
```
