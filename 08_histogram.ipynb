# Histogram
A histogram is a graph that shows how data is distributed.

It tells us how many values fall into ranges (groups).

```python
import matplotlib.pyplot as plt
import numpy as np

score = np.random.normal(loc= 80, scale= 10, size =100)
score = np.clip(score,0, 100)

plt.hist(score, bins=10,
         color="lightblue",
         edgecolor = "black")
plt.title("Exam Score")
plt.xlabel("Score")
plt.ylabel("No of Students")
plt.show()
```
