# pie()
 is used to create a pie chart.

A pie chart shows data as parts of a whole (like a circle cut into slices).

```python
import matplotlib.pyplot as plt


food = ["Rice", "Fruits", "Vegetables", "Meat"]
amount = [30, 20, 40, 10]
Colors = ["#d3d3d3","yellow", "green","red"]
plt.pie(amount, labels=food,
        autopct="%1.1f%%",
        colors= Colors)
plt.title("Daliy Consumption", fontweight = "bold")
plt.show()
```

# explode
 is used to separate (pull out) one or more slices of a pie chart.

```python
import matplotlib.pyplot as plt


food = ["Rice", "Fruits", "Vegetables", "Meat"]
amount = [30, 20, 40, 10]
Colors = ["#d3d3d3","yellow", "green","red"]
plt.pie(amount, labels=food,
        autopct="%1.1f%%",
        colors= Colors,
        explode=[0,0,0,0.1],
        shadow=True)
plt.title("Daliy Consumption", fontweight = "bold")
plt.show()
```

# Start Angle
As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0:

```python
import matplotlib.pyplot as plt


food = ["Rice", "Fruits", "Vegetables", "Meat"]
amount = [30, 20, 40, 10]
Colors = ["#d3d3d3","yellow", "green","red"]
plt.pie(amount, labels=food,
        autopct="%1.1f%%",
        colors= Colors,
        explode=[0,0,0,0.1],
        shadow=True,
        startangle=90)
plt.title("Daliy Consumption", fontweight = "bold")
plt.show()
```

# legend() 
is used to show a label box (key) on a graph.

It explains:
which line or color belongs to which data

```python
import matplotlib.pyplot as plt


food = ["Rice", "Fruits", "Vegetables", "Meat"]
amount = [30, 20, 40, 10]
Colors = ["#d3d3d3","yellow", "green","red"]
plt.pie(amount, labels=food,
        autopct="%1.1f%%",
        colors= Colors,
        explode=[0,0,0,0.1],
        shadow=True,
        startangle=90)
plt.title("Daliy Consumption", fontweight = "bold")
plt.legend()
plt.show()
```
