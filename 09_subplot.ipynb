# SubPlot
multiple graphs inside one figure/window.

Instead of opening separate graph windows, you place graphs together.

```python
import matplotlib.pyplot as plt

#First graph
plt.subplot(1,2,1)
x1 = [1,2,3,4,5,6]
y1 = [0,2,4,6,8,10]
plt.plot(x1,y1)
plt.title("Graph1")

# Second graph

y2 = [1,3,5,7,11,13]
plt.subplot(1,2,2)
plt.plot(x1,y2)
plt.title("Graph2")

plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,8,16,32]
figure, axis = plt.subplots(2,2)
axis[0,0].plot(x, y, color = "red")
axis[0,0].set_title("2x")

axis[0,1].plot(x, [i*2 for i in x], color = "green")
axis[0,1].set_title("x*2")

axis[1,0].plot(x, [i*3 for i in x], color = "blue")
axis[1,0].set_title("x3")

axis[1,1].plot(x, [i*4 for i in x], color = "yellow")
axis[1,1].set_title("x4")

plt.show()
```

# tight_layout() 
automatically fixes spacing between subplots.

It prevents graphs, titles, and labels from overlapping.

```python
x = [1,2,3,4,5]
y = [2,4,8,16,32]
figure, axis = plt.subplots(2,2)
axis[0,0].plot(x, y, color = "red")
axis[0,0].set_title("2x")

axis[0,1].plot(x, [i*2 for i in x], color = "green")
axis[0,1].set_title("x*2")

axis[1,0].plot(x, [i*3 for i in x], color = "blue")
axis[1,0].set_title("x3")

axis[1,1].plot(x, [i*4 for i in x], color = "yellow")
axis[1,1].set_title("x4")

plt.tight_layout()

plt.show()
```

# Differenct Charts

```python
x = [1,2,3,4,5]
y = [2,4,8,16,32]
figure, axis = plt.subplots(2,2)
axis[0,0].plot(x, y, color = "red")
axis[0,0].set_title("2x")

axis[0,1].plot(x, [i*2 for i in x], color = "green")
axis[0,1].set_title("x*x")

axis[1,0].bar(x, [i*3 for i in x], color = "blue")
axis[1,0].set_title("x3")

axis[1,1].plot(x, [i*4 for i in x], color = "yellow")
axis[1,1].set_title("x4")

plt.tight_layout()
plt.show()
```

# Super Title
You can add a title to the entire figure with the suptitle() function:

```python
x = [1,2,3,4,5]
y = [2,4,8,16,32]
figure, axis = plt.subplots(2,2)
axis[0,0].plot(x, y, color = "red")
axis[0,0].set_title("2x")

axis[0,1].plot(x, [i*2 for i in x], color = "green")
axis[0,1].set_title("x*x")

axis[1,0].bar(x, [i*3 for i in x], color = "blue")
axis[1,0].set_title("x3")

axis[1,1].plot(x, [i*4 for i in x], color = "yellow")
axis[1,1].set_title("x4")

plt.tight_layout()
plt.suptitle("Figures", fontweight= "bold")
plt.show()
```
