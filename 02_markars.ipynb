# Markers
Markers are symbols used to show individual data points on a graph.

Instead of only a line, you can show points like:

dot (•)

circle (o)

star (*)

square (s)

These symbols are called markers.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [4, 6, 5]

plt.plot(x, y, marker='s')
plt.show()
```

# markersize
Use to increase or decrease markersize

```python
import matplotlib.pyplot as plt

x = [1, 2, 3,4,5]
y = [4, 6, 5,9,2]

plt.plot(x, y, marker='.', markersize = 30)
plt.show()

# Shortcut for markersize is ms
```

# markerfacecolor
 is used to set the inside color of a marker in a graph.

It only affects the filled part inside the marker, not the line.

Marker = the point shape (circle, star, etc.)

Marker face color = the inside color of that shape

```python
import matplotlib.pyplot as plt

x = [3, 2, 9,4,5]
y = [2, 6, 5,9,2]

plt.plot(x, y, marker='.', markersize = 20, markerfacecolor = "#3d3a3a")

# Shortcut for markerfacecolor is mfc
```

# markeredgecolor
 is used to set the border (outline) color of a marker in a graph.

It controls the outside ring of the point.

```python
import matplotlib.pyplot as plt

y = [1,9,5,6,2,8]
plt.plot(y,marker =".", ms = 20, markerfacecolor = "#091b9e", markeredgecolor = "#0f0e0e")
plt.show()

# shorcut for markeredgecolor is mec
```

# linestyle
 is used to control how the line looks in a graph

| Style  | Meaning              |
| ------ | -------------------- |
| `'-'`  | solid line (default) |
| `'--'` | dashed line          |
| `':'`  | dotted line          |
| `'-.'` | dash-dot line        |
|`'none'`| no line              |
|`'solid'` | solid line           |


```python
import matplotlib.pyplot as plt

y = [1,9,5,6,2,8]
plt.plot(y,marker =".", ms = 20, markerfacecolor = "#091b9e", markeredgecolor = "#0f0e0e", linestyle = '-.')
plt.show()

#shortcut for linestyle is ls
```

# linewidth
linewidth controls the thickness of the line in a graph.

```python
import matplotlib.pyplot as plt

y = [1,9,5,6,2,8]
plt.plot(y,marker =".", ms = 20, markerfacecolor = "#091b9e", markeredgecolor = "#0f0e0e",
          linestyle = 'solid', linewidth = 2)
plt.show()
```

# color
color is used to set the color of the line

```python
import matplotlib.pyplot as plt

y = [1,9,5,6,2,8]
plt.plot(y,marker =".", ms = 20, markerfacecolor = "#091b9e", markeredgecolor = "#0f0e0e",
          linestyle = 'solid', linewidth = 2, color = '#610606')
plt.show()
```

# Coloring Two different lines

```python
# First method

import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1,
         marker =".", 
         ms = 20, 
         markerfacecolor = "#091b9e",
           markeredgecolor = "#0f0e0e",
          linestyle = 'solid',
            linewidth = 2,
              color = '#610606'
        )

plt.plot(x2,y2,marker = ".", ms = 20,
        markerfacecolor = "#ff7f0e",
        markeredgecolor = "#0f0e0e",
        linestyle = 'solid',
        linewidth = 2,
        color = 'green'
        )
plt.show()
```

# 2nd Method
If the all the properties are same then you can do this like:

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

lineStyle = dict(
            marker = ".", ms = 20,
            markerfacecolor = "#ff7f0e",
            markeredgecolor = "#0f0e0e",
            linestyle = 'solid',
            linewidth = 2,
            color = 'green')
plt.plot(x1, y1, **lineStyle)
plt.plot(x2, y2, **lineStyle)
plt.show()
```
