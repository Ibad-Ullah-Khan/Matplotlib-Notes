# Matplotlib
Matplotlib is a Python library used to create graphs and charts.

```python
from matplotlib import pyplot as plt
```

# 1. import
import means:

“Bring code from another library into my program.”

Python itself does not know how to make graphs.
So we import Matplotlib.
# 2. pyplot
This is a module inside Matplotlib.

matplotlib → whole library

pyplot → part used for plotting graphs

pyplot contains functions like:

plot(),
bar(),
show(),
title(),
# 3. as plt
This creates a short nickname.

# plt.plot()
The plot() function is used to draw points (markers) in a diagram.

plot() does NOT just show dots, It automatically connects them with a line
# plt.show()
show() displays the graph on the screen.

Without it, the graph may not appear.

```python
import numpy as np

x = np.array([10,11,12,13,14])
y = np.array([10,20,40,80,160])

plt.plot(x,y)
plt.show()
```

# Plotting Without Line
To plot only the markers, you can use shortcut string notation parameter 'o', which means 'rings'.

```python
x = np.array([10,11,12,13,14])
y = np.array([10,20,40,80,160])

plt.plot(x,y,"o")
plt.show()
```

# Default X-Points
If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

```python
import matplotlib.pyplot as plt

y = [1,9,5,6,2,8]
plt.plot(y)
plt.show()
```
